### 4. Average duration of filmes grouped by color

```python
pd.crosstab(movies.genres, movies.color,values=movies['duration'],aggfunc='mean')
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
movies['films_like']=movies['actor_1_facebook_likes']+movies['actor_3_facebook_likes']+movies['actor_2_facebook_likes']
```

```python
pop_directors=pd.crosstab(movies.director_name,columns='likes',values=movies.films_like,aggfunc="mean")
```

```python
pop_directors.sort_values(by='likes',ascending=False)
```
