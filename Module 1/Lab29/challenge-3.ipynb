Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
agg_spent = orders.groupby('CustomerID', as_index=False).agg({'amount_spent':'sum'})
```

```python
agg_spent
```

```python
agg_spent.quantile(0.95)
```

```python
#Why does it return a "ValueError: Can only compare identically-labeled Series objects" if I compare to the formula and not the float output
vip_filt=agg_spent['amount_spent']>=5840.182
vip=agg_spent[vip_filt]
vip.insert(1,'Label','VIP')

print(agg_spent.quantile(0.75))
pref_filt=(agg_spent['amount_spent']>=1661.64) & (agg_spent['amount_spent']<5840.182)
pref=agg_spent[pref_filt]
pref.insert(1,'Label','Preferred')
```

```python
vip
```

```python
pref
```

```python
top_customers=vip.append(pref)
```

```python
top_customers
```

```python
label_cust=top_customers.drop(columns='amount_spent')
```

```python
country_cust=orders[['CustomerID','Country']].drop_duplicates()
```

```python
country_cust
```

```python
final=pd.merge(country_cust,label_cust,how='left',on='CustomerID').dropna().reset_index().sort_values('Label')
```
