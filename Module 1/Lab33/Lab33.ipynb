```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\lenovo\Downloads\example_data_cleaning.csv')
```

```python
#I first check if there are duplicate transaction records
df.sort_values('TransactionID')
#There are not.
```

```python
df.sort_values('ClientID')
#ClientID can be duplicate. However, I see that some info are inconsistent. For example, the same client does not have 
#always the same birth date. 
```

```python
df.loc[22,'BirthYear']=1967
df.loc[21,'BirthYear']=1999
df.loc[1,'BirthYear']=1967
```

```python
df.sort_values('ClientID')
```

```python
#In the case of client 34989, there are only 2 records with different BirthYear. I can't judge which one is correct.
#I will therefore replace both years with a np.nan to show the values are missing and need rechecking.
df.loc[18,'BirthYear']=np.nan
df.loc[19,'BirthYear']=np.nan
```

```python
df
```

```python
#Then I need to unify professions.
df['Profession'].unique()
```

```python
df['Profession'].replace({'etudient':'student','Student':'student','BDM':'bdm','HR':'hr','Manager':'manager'},inplace=True)
```

```python
df['Profession'].unique()
```

```python
#I can then see that the department has a unique value, and therefore is not a relevant piece of information for analysis.
#Let's drop it!
df.drop(columns='Department',inplace=True)
```

```python
#After that, I can check in transaction amount if there is anything weird.
df.describe()
```

```python
#I can see the average amount of transactions is 101009 euros.
#Yet, the max is 2.3M euros. It must be an outlier. Let's delete this row so our insights are not blurred.
df.drop(index=10,inplace=True)
```

```python
df
```

```python
#Now, to study the correlation between say birth year and risk, or profession and risk, we may want to encode
#Risk values and Profession values as numeric data
encoded=pd.get_dummies(df,['Profession','Risk'])
```
